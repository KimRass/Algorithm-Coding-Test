```python
from collections import deque

N = int(input())

# arr = list()
arr = deque()
cnt = 0
def dfs(row):
    global cnt
    
    if row == N:
        cnt += 1
    else:
        for i in range(N):
            for j in range(row):
                if i in [arr[j] + j - row, arr[j], arr[j] - j + row]:
                    break
            else:
                arr.append(i)
                dfs(row + 1)
                arr.pop()

dfs(0)
print(cnt)
```
