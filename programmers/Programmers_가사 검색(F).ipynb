## Trie Data Structure

```python
class Node():
    def __init__(self, key, flag=None):
        self.key = key
        self.flag = flag
        self.child = dict()
    
class Trie():
    def __init__(self):
        self.head = Node(None)
    def insert(self, word):
        cur_node = self.head
        for char in word:
            if char not in cur_node.child:
                cur_node.child[char] = Node(char)
            cur_node = cur_node.child[char]
        cur_node.flag = True
    def search(self, word):
        cur_node = self.head
        leng = len(word)
        if word[-1] == "?":
            for i, char in enumerate(word):
                if char == "?":
                    break
                elif char in cur_node.child:
                    cur_node = cur_node.child[char]
                else:
                    return 0
            cur_nodes = [cur_node]
            next_nodes = list()
            cnt = 0
            for _ in range(leng - i):
                for node in cur_nodes:
                    next_nodes.extend(list(node.child.values()))
                cur_nodes = next_nodes
                next_nodes = list()
            for node in cur_nodes:
                if node.flag == True:
                    cnt += 1
            return cnt
        else:
            cur_node = self.head
            cur_nodes = [cur_node]
            next_nodes = list()
            for i, char in enumerate(word):
                if char != "?":
                    break
                else:
                    for node in cur_nodes:
                        next_nodes.extend(list(node.child.values()))
                    cur_nodes = next_nodes
                    next_nodes = list()
            cnt = 0
            for cur_node in cur_nodes:
                # print(cur_node.key)
                for char in word[i:]:
                    if char in cur_node.child:
                        cur_node = cur_node.child[char]
                        # print(cur_node.key)
                    else:
                        break
                else:
                    if cur_node.flag == True:
                        cnt += 1
                        # print(cur_node.key + "!")
            return cnt
        
def solution(words, queries):
    trie = Trie()
    for word in words:
        trie.insert(word)
    # for query in queries:
    #     trie.search(query)
    return [trie.search(query) for query in queries]
    # return None
```

## Regular Expression

```python
import re

def solution(words, queries):
    words = " " + "  ".join(words) + " "
    # print(words)
    query2cnt = dict()
    ans = list()
    for query in queries:
        if query in query2cnt.keys():
            ans.append(query2cnt[query])
        else:
            try:
                chars = re.search(r"[a-z]+", query).group()
            except:
                pass
            n_wc = query.count("?")
            if query[0]=="?" and query[-1]=="?":
                cnt = len(re.findall(rf"[ ][a-z]{{{n_wc}}}[ ]", words))
                ans.append(cnt)
                query2cnt.update({query:cnt})
            elif query[0]=="?":
                cnt = len(re.findall(rf"[ ][a-z]{{{n_wc}}}{chars}[ ]", words))
                ans.append(cnt)
                query2cnt.update({query:cnt})
            elif query[-1]=="?":
                cnt = len(re.findall(rf"[ ]{chars}[a-z]{{{n_wc}}}[ ]", words))
                ans.append(cnt)
                query2cnt.update({query:cnt})
    return ans
```
