# Using `itertools.product`

```python
from itertools import product

N, M = list(map(int, input().split()))

string = "range(1, N + 1), "
for i in eval("product(" + (string*M)[:-2] + ")"):
    for j in i:
        print(j, end=" ")
    print()
```

# Using Backtracking

```python
N, M = list(map(int, input().split()))

arr = list()
def dfs(step):
    if step == M:
        print(*arr, sep=" ")
    else:
        for i in range(1, N + 1):
            arr.append(i)
            dfs(step + 1)
            arr.pop()

dfs(0)
```
